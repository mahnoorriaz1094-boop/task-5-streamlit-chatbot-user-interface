## Step 6 — Push to GitHub

```python
import subprocess

# ── FILL THESE IN ──────────────────────────────────────────
GITHUB_USERNAME = 'your_github_username'
GITHUB_TOKEN    = 'your_personal_access_token'
REPO_NAME       = 'your_repo_name'
BRANCH          = 'main'
# ───────────────────────────────────────────────────────────

REPO_URL  = f'https://{GITHUB_USERNAME}:{GITHUB_TOKEN}@github.com/{GITHUB_USERNAME}/{REPO_NAME}.git'
CLONE_DIR = f'/content/{REPO_NAME}'

def run(cmd):
    r = subprocess.run(cmd, shell=True, capture_output=True, text=True)
    if r.stdout: print(r.stdout)
    if r.stderr: print(r.stderr)

run(f'git clone {REPO_URL} {CLONE_DIR}')

# Copy app files
run(f'cp /content/app.py {CLONE_DIR}/')
run(f'cp /content/coretech_faq.csv {CLONE_DIR}/')
run(f'cp /content/CoreTech_Streamlit_App.ipynb {CLONE_DIR}/')

# Copy screenshots if they exist
run(f'mkdir -p {CLONE_DIR}/screenshots')
run(f'cp /content/screenshots/*.png {CLONE_DIR}/screenshots/ 2>/dev/null || true')

run(f'git -C {CLONE_DIR} config user.email "{GITHUB_USERNAME}@users.noreply.github.com"')
run(f'git -C {CLONE_DIR} config user.name "{GITHUB_USERNAME}"')
run(f'git -C {CLONE_DIR} add .')
run(f'git -C {CLONE_DIR} commit -m "Task 5: Add CoreTech Streamlit web app"')
run(f'git -C {CLONE_DIR} push origin {BRANCH}')

print('\n✅ Pushed to GitHub!')
```
